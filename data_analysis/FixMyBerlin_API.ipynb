```python
import pandas as pd
import numpy as np
import requests
```

# Fix My Berlin API

See https://fixmyberlin.de/api

```python
url = "https://fixmyberlin.de/api/v1/sections"
json = requests.get(url).json()
results = pd.DataFrame(json.get('results'))
```

```python
pd.DataFrame(results.iloc[0]['details'])
```

```python
happy_bike_idx_dict = {}
happy_bike_idx_list = []
for index, row in results.iterrows():
    happy_bike_idx = pd.DataFrame(row['details'])['happy_bike_index'].mean()
    borough = row['borough'] + " " + row['street_name']
    happy_bike_idx_dict[borough] = happy_bike_idx
    happy_bike_idx_list.append((borough, happy_bike_idx))
```
