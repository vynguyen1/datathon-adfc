```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import scipy.stats as stats
import statsmodels.api as sm
import warnings
import datetime
import itertools
from pylab import rcParams

rcParams['figure.figsize'] = 18, 8
plt.style.use('fivethirtyeight')     # see plt.style.available to see other styles

warnings.simplefilter(action='ignore', category=FutureWarning)
pd.options.mode.chained_assignment = None  # default='warn'
warnings.filterwarnings("ignore")
```

```python
berlinFilenamesList = glob.glob('ADFC/Berlin*.csv')
list_of_dfs = []
for berlinFilename in berlinFilenamesList:
    berlin_data = pd.read_csv(berlinFilename, index_col=0)
    berlin_data['date'] = pd.to_datetime(berlin_data['date'])
    berlin_data.set_index('date', inplace=True)
    berlin_data = berlin_data.drop('timestamp', 1)
    berlin_data = berlin_data.resample('M').apply({'comptage':'sum'})
    berlin_data['group'] = berlinFilename
    list_of_dfs.append(berlin_data)
df = pd.concat(list_of_dfs)
```

## Forecasting - Prediction Model

```python
plt.style.use('bmh')
```

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
```

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(df_berlin_2,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit(full_output=False, disp=False)
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

ARIMA(1, 1, 1)x(1, 1, 1, 12)12 - AIC:964.9700674200337  
-> yields the lowest AIC value of 297.78 -> we consider this to be the optimal option

### Fitting the ARIMA model

```python
mod = sm.tsa.statespace.SARIMAX(df_berlin_5,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit(full_output=False, disp=False)
print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
pred = results.get_prediction(start=pd.to_datetime('2019-12-31'), dynamic=False)
pred_ci = pred.conf_int()
ax = df_berlin_5['2016':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Comptage')
plt.legend()
plt.show()
```

```python
pred_uc = results.get_forecast(steps=48)
pred_ci = pred_uc.conf_int()
ax = df_berlin_5.plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Comptage')
plt.legend()
plt.show()
```
